```python
# boxplot grid
def build_grid(y, nrows, ncols):
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(16, 16))
    for idx, ax in enumerate(axes.flat):
        sns.boxplot(x='price_range', y=y[idx], data=data, ax=ax)
```

```python
def add_ones(X):
    ones = np.ones((X.shape[0], X.shape[1] + 1))
    ones[:, :-1] = X
    return ones
```

```python
X, y = data.iloc[:, :-1].values, data['price_range'].values
X = add_ones(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
# scaling
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
class Perceptron:
    
    def __init__(self, param_dim, num_labels):
           self.params = np.random.randn(num_labels,
                                         param_dim)
            
    def predict(self, X):
          return np.argmax(self.params.dot(X), axis=0)
        
    def load_params(self, params):
        self.params = params
        
    def train(self, X, y, num_iters=100, X_val=None, y_val=None):
        best_accuracy = 0
        if X_val is None:
            X_val = X
            y_val = y
                
        for iter_num in range(num_iters):
            for i in range(len(X)):
                prediction = self.predict(X[i])
                if prediction != y[i]:
                    self.params[prediction] -= X[i]
                    self.params[y[i]] += X[i]
                    
            val_prediction = self.predict(X_val.T)
            iter_accuracy = np.sum(val_prediction == y_val) / len(val_prediction)
            if iter_accuracy > best_accuracy:
                best_accuracy = iter_accuracy
                cached_params = np.copy(self.params)
                
        return (best_accuracy, cached_params)
```

```python
param_dim = X_train.shape[1]
perceptron = Perceptron(param_dim=param_dim, num_labels=4)
```

```python
val_accuracy, params = perceptron.train(X_train, y_train, X_val=X_test, y_val=y_test, num_iters=2000)
```

```python
print('Best accuracy on validation set: ', val_accuracy)
```

```python
perceptron.load_params(params)
```

```python
train_pred = perceptron.predict(X_train.T)
train_accuracy = np.sum(train_pred == y_train) / len(train_pred)
```

```python
print('Accuracy on training set: ', train_accuracy)
```
