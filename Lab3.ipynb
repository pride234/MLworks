```python
X, Y = data.iloc[:, :-1], data['price_range']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
# scaling
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
def print_metrics(classifier, X, Y):
    prediction = classifier.predict(X)
    print(classification_report(Y, prediction))
```

```python
model = SVC()
model.fit(X_train, Y_train)
```

```python
print('Metrics for train set:')
print_metrics(model, X_train, Y_train)
print('Metrics for test set:')
print_metrics(model, X_test, Y_test)
```

```python
param_grid = {'C': [0.1,1, 10, 100, 1000], 'gamma': [1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf', 'linear']} 
```

```python
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=3)
```

```python
grid.fit(X_train, Y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
print('Metrics for train set:')
print_metrics(grid, X_train, Y_train)
print('Metrics for test set:')
print_metrics(grid, X_test, Y_test)
```
